# 1. Encabezados 

Para crear un encabezado, al inicio de la línea debemos agregar el símbolo de octothorpe (#) "gatito" o #número". Toma en consideración que el número de símbolos que uses, corresponderá al nivel jerárquico del encabezado.

Por ejemplo: 
# Encabezado de nivel 1
## Encabezado de nivel 2
### Encabezado de nivel 3
#### Encabezado de nivel 4
##### Encabezado de nivel 5
###### Encabezado de nivel 6

# 4. Énfasis 

## 4.1 Negritas

Para poner una palabra o texto en negritas, es necesario que utilices dos asteriscos o guiones bajos antes  y después de la palabra o frase.

Por ejemplo:<br>
**Lorem ipsum** dolor sit amet consectetur, __adipiscing__ elit rutrum commodo. Non tristique erat habitant sed risus mollis inceptos mi hendrerit nascetur, rhoncus porttitor nisi curae eu pretium ultrices magnis tempus consequat fermentum, **dictum litora dis dignissim pharetra cras lectus facilisis quis**.

## 4.2 Itálica o cursiva
Para que un texto se encuentre con una tipografía cursiva, se requiere agregar un asterisco o guión bajo, antes y después de la palabra o frase.

Por ejemplo: <br>
*Velit* ante nec dignissim ac suspendisse lobortis libero nisl, elementum in parturient posuere orci id mi, *rhoncus pharetra sapien hendrerit cum ultrices porta. Habitasse nunc senectus etiam pharetra fames urna a platea nascetur dignissim duis*, curabitur risus laoreet luctus dictumst elementum ultrices tincidunt faucibus. Natoque erat orci _faucibus litora_ phasellus nibh aliquam in et, aliquet cursus eleifend dignissim hendrerit ut sollicitudin metus.

Para crear una cita, agrega el símbolo de mayor que (>) al inicio del párrafo. Ejemplo:

Pellentesque turpis scelerisque fames ligula inceptos velit arcu, egestas netus cras maecenas penatibus interdum lacinia, hendrerit ante nulla nascetur tempus vel. Volutpat aliquam eget suscipit in et phasellus odio turpis sem, augue non elementum dui etiam primis ut auctor.

Integer mi nec curabitur consequat ornare libero nulla, ullamcorper quis habitasse dignissim egestas mattis elementum, phasellus sollicitudin nibh duis turpis fames.

### 4.3.1 Citas con múltiples párrafos
Las citas pueden tener múltiples párrafos, para ello, es necesario agregar el símbolo mayor que (>) en la línea en blanco que separa los párrafos. Ejemplo:

> Este es el primer párrafo.
>
> Este es el segundo párrafo.

### 4.3.2 Citas anidadas
Las citas pueden anidarse mediante el uso de >> al inicio del párrafo que deseamos anidar.

> Este es el primer párrafo.
>
>> y este es párrafo anidado.

### 4.3.3 Citas con otros elementos
En las citas se pueden usar otros elementos de Markdown; sin embargo, toma en consideración que no todos los elementos pueden llegar a usarse, en este caso, te recomiendo que experimentes para que identifiques cuales elementos si pueden usarse. Ejemplo:

> ##### Resultados del análisis:
>
> - Estadístico.
> - Numérico.
>
> En el análisis estadístico se calculó la **media** para...

# 5. Listas 
### 5.1 Listas ordenadas
Para crea una lista ordenada, agrega un número con un punto por cada uno de los elementos de la lista. Cabe mencionar que los números pueden no seguir una orden numérico, pero la listas siempre deberán empezar con el número 1. Ejemplo:

1. Elemento 1.
2. Elemento 2.
3. Elemento 3.
5. Elemento 4.
7. Elemento 5.

### 5.2 Lista anidada
Una lista anidada puede crearse usando una sangría de 4 espacios o una tabulación y después, agregamos un número tal como se realiaz en una lista ordenda. Ejemplo:
1. Elemento 1
2. Elemento 2
3. Elemento 3
    1. Elemento anidado 1
    2. Elemento anidado 2
    3. Elemento anidado 3
4. Elemento 4
5. Elemento 5

### 5.3 Lista no numerada 
Agrega un guión bajo (-), asterisco (*), o el signo positivo (+) al inicio del elemento. Ejemplo:

* 1er. elemento
+ 2do. elemento
- 3er. elemento

### 5.4 Lista no numerada anidada
Agrega un guión bajo (-), asterisco (*), o el signo positivo (+) al inicio del elemento y los elementos anidados deberán iniciar con 4 espacios o un tabulador, seguido de un (-), (+) o un asterisco. Ejemplo:

+ Elemento 1
+ Elemento 2
+ Elemento 3
    - Elemento anidado 1
    - Elemento anidado 2
    - Elemento anidado 3
- Elemento 4
- Elemento 5

### 5.5 Agregar elementos dentro de una lista
Para incluir otro elemento dentro de una lista y preservar la continuidad de la lista, al inicio del elemento usamos un guión bajo (-) o (+) seguido de 4 espacios o un tabulador, y para indicar un párrafo debajo del elemento previo, solo usamos una sangría de 4 espacios o un tabulador al inicio de la línea. Por ejemplo:

-    Reporte de ventas diarias.
-    Reporte de ventas mensuales.
    El reporte de ventas mensuales solo incluye los 3 primeros meses del año.
    y el cierre se realiza los días 28 del mes.
- Reporte de venta anuales.

### 5.6 Agregar elementos dentro de una lista
Para incluir otro elemento dentro de una lista y preservar la continuidad de la lista, al inicio del elemento usamos un guión bajo (-) o (+) seguido de 4 espacios o un tabulador, y para indicar una cita debajo del elemento previo, usamos una sangría de 5 espacios al inicio de la línea y posteriormente el símbolo de (>). Por ejemplo:

*    Reporte de ventas diarias.
*    Reporte de ventas mensuales.
     >   El reporte de ventas mensuales solo incluye los 3 primeros meses del año.
     >   y el cierre se realiza los días 28 del mes.
*    Reporte de venta anuales.

# 6. Imágenes 

Si tu texto requiere la inserción de una imagen, agrega un signo de admiración (!) seguido de corchetes [] y unos paréntesis(). Dentro de los corchetes se proporciona el texto del titulo de la imagen y dentro del paréntesis debes de proporcionar la ruta de acceso del archivo que contiene la imagen. Se recomienda que el archivo con la imagen se encuentre en la misma carpeta donde se ubica tu archivo de Jupyter. Ejemplo:

![](logoPython.jpg)

# 8. Tablas
En el caso de ser requerida una tabla, en Markdown establecemos la primer línea de encabezados, donde cada columna será delimitada por el símbolo (|), en la siguiente línea y para cada una de las columnas usaremos tres guiones bajos para crear la línea de la tabla (---)

| Nombre | Apellido | Nivel |
| --- | --- | --- |
| Gabriel | Ruiz | A1 |
| Joe | Doe | A2 |
| Jane | Doe | A3 |

En algunas ocasiones puede llegar a ser un poco complicado y tedioso crear tablas en Markdown, por tal razón, te recomiendo que utilices la página web [Markdown Tables Generator](https://www.tablesgenerator.com/markdown_tables "Generador de Tablas en Markdown") para construir tu tabla a través de la interfase gráfica, copia el código que se genera en la página web y pegalo en tu documento.

Para alinear las columnas a la izquierda, derecha o centrado, agrega el símbolo de dos puntos en la línea de los (---). Si quieres alineación a la izquiera coloca los (:) antes de los tres guiones; si deseas el texto centrado, delimita los tres guiones con los dos puntos y si deseas alinear el texto a la derecha, coloca los dos puntos al final de los tres puntos. Ejemplo:

| Nombre | Apellido | Equipo |
| :--- | :---: | ---: |
| Michael | Jordan | Chicago Bulls |
| Dan | Marino | Miami Dolphins |
| Hugo | Sánchez | Real Madrid |
