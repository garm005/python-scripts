```python
# Especifcando los integrantes de los equipos
team_a = {'Peach', 'Mario', 'Luigi', 'Toad'}
team_b = {'Yoshi', 'Toad', 'Peach', 'Donkey'}
team_c = {'Mario', 'Yoshi', 'Peach', 'Dixie'}

# Vamos a realizar la union de los conjuntos para encontrar los nombres que son unicos 
all_characters = team_a.union(team_b).union(team_b)
print(f'Los competidores son: {all_characters}')
print(f'Los nombres de los competidores se encuentran almacenados en un :{type(all_characters)}\n')

print(zip('abc', [team_a, team_b, team_c]))


# Iteracion, donde cada personaje que se encuentre en el conjunto de todos los personajes se analizará
for personaje in all_characters:
    # Creando una lista vacia. En esta lista se almacenará de manera temporal si el personaje se encuentra en 1 o mas equipos
    in_team = []
    print(personaje)

    # Vamos a  etiquetar cada equipo con las letras a, b, c
    for ID, team in zip('abc', [team_a, team_b, team_c]):
        print(ID)
        print(team)
        print(f'El equipo se encuentra almacenado en un :{type(team)}\n')
        
        # Vamos a verificar que el personaje se encuentre en el equipo. Si el personaje se encuentran en el equipo, que agregue
        # en la lista el equipo donde se identifica
        if personaje in team:
            in_team.append(ID)
            print(team)
    
    # El metodo join toma todos los elementos que pueden iterarse y los une en un solo string
    label_team = '-'.join(in_team)
    
    # Utilizando una expersion ternaria
    p = 's' if len(in_team) > 1 else ''

    # Visualizando el resultado
    print(f'{personaje} en team{p}: {label_team}')
```
